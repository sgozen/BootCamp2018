```python
""" Problem 1.1 """

import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
iris = data('iris')
iris.sample(5)
#iris[iris['Species']=='setosa'].plot(kind='scatter', x ='Petal.Length',y ='Petal.Length')


colors = ['green', 'black', 'red']

fig, ax = plt.subplots()

for spec, color in zip(iris['Species'].unique(), colors):
    iris[iris['Species'] == spec].plot(kind='scatter', x='Petal.Length', y='Sepal.Width', ax=ax, c=color, label=str(spec))
plt.show()


fig, ax = plt.subplots()

for spec, color in zip(iris['Species'].unique(), colors):
    iris[iris['Species'] == spec].plot(kind='scatter', x='Petal.Length', y='Sepal.Length', ax=ax, c=color, label=str(spec))
plt.show()

iris['Species'].unique()
```

```python
""" Problem 1.2 """

poisons = data('poisons')
poisons.boxplot(by='poison')
plt.ylabel('Time of Survival')
plt.show()
# Poison 3 is the most dangerous.
poisons.groupby(['poison', 'treat']).mean().unstack().plot(kind='bar')
```

```python
""" Problem 1.3 """
diamonds = data('diamonds')
diamonds.sample(5)
diamonds.groupby('cut')['price'].mean().plot(kind='bar')
plt.title('Diamond price by cut')
plt.show()

diamonds.groupby('color')['price'].mean().plot(kind='bar')
plt.title('Diamond price by color')
plt.show()


diamonds['volume'] = diamonds['x'] * diamonds['y'] * diamonds['z']
diamonds[diamonds.color == 'H'].groupby('cut')['volume'].mean()[['Fair','Ideal']].plot(kind='bar')
plt.title('Volumes of Diamond')
plt.show()
## Volume might be the important factor for explaning...
```
