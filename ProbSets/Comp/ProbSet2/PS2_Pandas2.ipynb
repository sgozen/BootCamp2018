```python
"Problem 1.1"

import pandas as pd
from matplotlib import pyplot as plt
from pydataset import data
Arbuthnot = data("Arbuthnot")
Arbuthnot
plt.title("Arbuthnot")
plt.plot(Arbuthnot['Year'],Arbuthnot['Males'])
plt.plot(Arbuthnot['Year'],Arbuthnot['Females'])
plt.xlabel('Year')
plt.legend(['Males','Females'])
# After mass deaths due to plague, a number of people converted to Christians!
plt.show()
```

```python
"Problem 1.2"

Trees = data("trees")
Trees
plt.plot(Trees['Girth'],Trees['Height'])
plt.plot(Trees['Girth'],Trees['Volume'])
plt.title("Trees")
plt.xlabel('Girth')
plt.legend(['Height','Volume'])
plt.show()
#Greater the number of the Girth, Higher the Height and Volume the Trees grow.
```

```python
"Problem 1.3"

Roads = data("road")
Roads
plt.scatter(Roads['drivers'],Roads['deaths'])
plt.scatter(Roads['drivers'],Roads['popden'])
plt.title('Road')
plt.xlabel('Number of Drivers')
plt.legend(['deaths','popden'])
plt.show()

#Greater the number of drivers, the greater number of death.
#The number of driver does not seem to relate to population density.
```

```python
"Problem 1.4"
BF = data("bfeed")
BF
plt.title('Breast Feed Duration among Ages')
plt.scatter(BF['agemth'],BF['duration'])
plt.xlabel('Age')
plt.ylabel('Feed Duration')
plt.legend(['duration'])
plt.show()

plt.scatter(BF['race'],BF['duration'])
plt.title('Breast Feed Duration among Races')
plt.xlabel('Races')
plt.ylabel('Feed Duration')
plt.legend(['duration'])
plt.show()

plt.show
#On average, mother with age between 18 to 24 has higher chance to breast feed longer than mothers of other ages.
#On average, people with Race 2 seems to breast feed their children with less amount of time.
```

```python
"Problem 1.5"
LG = data("lung")
plt.title('Survival Time and Ages')
plt.scatter(LG['time'],LG['age'])
plt.xlabel("Time of Survival")
plt.ylabel("Ages")
plt.legend(['age'])
plt.show()

plt.title('Calorie and Weight Loss')
plt.scatter(LG['meal.cal'], LG['wt.loss'])
plt.xlabel("Calorie Per Meal")
plt.ylabel("Weights Loss")
plt.legend(['Weight Loss'])
plt.show()

#Time of survival does not seem to have postive relationship with ages.
#Amoung patients, the calorie per meal does not seem to have postive relationship with the weight they lose.
```
